```python
Allcategories = ['Clothing', 'Footwear', 'Outerwear', 'Accessories']

# Mencari data dengan Purchase Amount (USD) sebesar 100 dan salah satu dari kategori pilihan
filterCat = df[(df['Purchase Amount (USD)'] == 100) & (df['Category'].isin(Allcategories))]
filterCat.sort_values(by='Category')
```

```python
category_counts = filterCat['Category'].value_counts()
# Mengurutkan tampilan kategori dari terbanyak ke terkecil
category_counts.sort_values(ascending=False)
```

```python
#['Clothing', 'Footwear', 'Outerwear', 'Accessories']
filclot=filterCat[(filterCat['Category'] == 'Clothing')]
sns.countplot(x="Subscription Status", data=filclot)
plt.show()

#Subscription Status(status berlangganan) ini mempengaruhi penjualan
```

```python
#dicount mempengaruhi jumlah terjual 
filclot=filterCat[(filterCat['Category'] == 'Clothing')]
sns.countplot(x="Discount Applied", data=filclot)
plt.show()

#Discount Applied  and Promo Code Used mempengaruhi tingkat penjualan
```
